```python
import requests
import json
import csv
import pandas as pd
import matplotlib.pyplot as plt

# 假设你已经有了API密钥和API端点
api_key = 'your_api_key'
api_endpoint = 'https://api.eastmoney.com/your_endpoint'  # 替换为实际的API端点

# 股票列表
stocks = ['000001.SZ', '600000.SH', '000002.SZ']  # 替换为实际的股票代码

# 发送HTTP请求并获取数据
def fetch_stock_data(stock_code):
    params = {
        'api_key': api_key,
        'stock_code': stock_code,
        # 其他必要的参数
    }
    response = requests.get(api_endpoint, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return None

# 存储数据的列表
data = []

# 遍历股票列表并获取数据
for stock in stocks:
    stock_data = fetch_stock_data(stock)
    if stock_data:
        # 假设返回的JSON数据中有'pe'和'pb'字段表示市盈率和市净率
        pe = stock_data.get('pe', 'N/A')
        pb = stock_data.get('pb', 'N/A')
        data.append({
            'Stock Code': stock,
            'PE Ratio': pe,
            'PB Ratio': pb
        })

# 将数据转换为DataFrame
df = pd.DataFrame(data)

# 保存到CSV文件
csv_file_path = 'stock_data.csv'
df.to_csv(csv_file_path, index=False)

# 可视化数据
df.set_index('Stock Code').plot(kind='bar', figsize=(10, 5))
plt.title('PE and PB Ratios of Selected Stocks')
plt.ylabel('Ratios')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd
from datetime import datetime
import time
import json

# 配置参数
stock_list = ['600519', '000001', '300750']  # 贵州茅台、平安银行、宁德时代
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36',
    'Referer': 'http://quote.eastmoney.com/'
}

def fetch_eastmoney_data(stock_code):
    """获取东方财富实时估值数据"""
    url = f'http://push2.eastmoney.com/api/qt/stock/get'
    params = {
        'ut': 'fa5fd1943c7b386f172d6893dbfba10b',
        'invt': 2,
        'fltt': 2,
        'fields': 'f43,f44,f45,f46,f60,f86,f169,f170,f152,f47,f48,f49,f50,f57,f58,f85,f116,f117,f161,f171,f172,f265,f292',
        'secid': f'1.{stock_code}' if stock_code.startswith('6') else f'0.{stock_code}'
    }
    
    try:
        response = requests.get(url, params=params, headers=headers)
        response.raise_for_status()
        data = json.loads(response.text)['data']
        
        return {
            '代码': stock_code,
            '名称': data.get('f58', 'N/A'),
            '最新价': data.get('f43', 0)/100,
            '市盈率(PE)': data.get('f162', 0)/100,
            '市净率(PB)': data.get('f167', 0)/100,
            '量比': data.get('f50', 0)/100,
            '更新时间': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        }
    except Exception as e:
        print(f"获取{stock_code}数据失败：{str(e)}")
        return None

def get_all_stocks_data():
    """批量获取股票数据"""
    all_data = []
    for code in stock_list:
        data = fetch_eastmoney_data(code)
        if data:
            all_data.append(data)
        time.sleep(1)  # 防止请求过于频繁
    return pd.DataFrame(all_data)

# 获取数据
df = get_all_stocks_data()

# 保存到CSV
if not df.empty:
    df.to_csv('stock_valuation.csv', index=False, encoding='utf_8_sig')
    print("数据已保存至 stock_valuation.csv")
else:
    print("未获取到有效数据")
```

```python
import requests
import pandas as pd
from datetime import datetime
import time

# ========== 配置参数 ==========
stock_codes = ['600519', '000001', '300750']  # 贵州茅台、平安银行、宁德时代
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36',
    'Referer': 'http://quote.eastmoney.com/'
}

# ========== 数据获取函数 ==========
def get_eastmoney_stock_data(code):
    """获取单只股票实时估值数据"""
    # 构造市场标识（沪市1. 深市0.）
    market = '1.' if code.startswith('6') else '0.'
    secid = market + code
    
    # 动态构造请求URL（通过浏览器开发者工具分析得出）
    url = "http://push2.eastmoney.com/api/qt/stock/get"
    params = {
        'ut': 'fa5fd1943c7b386f172d6893dbfba10b',
        'fltt': 2,
        'invt': 2,
        'fields': 'f58,f43,f57,f169,f170,f152,f46,f60,f84,f85,f116,f117,f47,f48,f86,f167,f164,f163,f92',
        'secid': secid
    }
    
    try:
        response = requests.get(url, params=params, headers=headers)
        response.raise_for_status()
        json_data = response.json()
        
        if json_data['data']:
            data = json_data['data']
            return {
                '股票代码': code,
                '股票名称': data.get('f58', 'N/A'),
                '最新价': data.get('f43', 0)/100,
                '市盈率(PE)': data.get('f162', 0)/100,
                '市净率(PB)': data.get('f167', 0)/100,
                '更新时间': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            }
        return None
    except Exception as e:
        print(f"获取{code}数据失败：{str(e)}")
        return None

# ========== 批量获取数据 ==========
def batch_fetch_data(codes):
    """批量获取股票数据"""
    all_data = []
    for code in codes:
        stock_data = get_eastmoney_stock_data(code)
        if stock_data:
            all_data.append(stock_data)
        time.sleep(1)  # 控制请求频率
    return pd.DataFrame(all_data)

# ========== 数据保存 ==========
def save_to_csv(df, filename=None):
    """保存数据到CSV"""
    if filename is None:
        filename = f"stock_valuation_{datetime.now().strftime('%Y%m%d')}.csv"
    
    df.to_csv(filename, 
             index=False, 
             encoding='utf_8_sig', 
             float_format='%.2f')
    print(f"数据已保存至 {filename}")

# ========== 执行主程序 ==========
if __name__ == "__main__":
    # 获取数据
    df = batch_fetch_data(stock_codes)
    
    if not df.empty:
        # 数据清洗
        df = df[['股票代码', '股票名称', '最新价', '市盈率(PE)', '市净率(PB)', '更新时间']]
        
        # 保存数据
        save_to_csv(df)
        
        # 展示数据
        display(df.style
               .format({'最新价': '{:.2f}', '市盈率(PE)': '{:.2f}', '市净率(PB)': '{:.2f}'})
               .background_gradient(subset=['市盈率(PE)'], cmap='RdYlGn_r')
               .background_gradient(subset=['市净率(PB)'], cmap='Blues'))
    else:
        print("未获取到有效数据")
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from matplotlib import rcParams
 
# 设置字体为支持中文的字体
rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
warnings.filterwarnings("ignore")  # 全局忽略所有警告

# 股票代码（小米集团-W，港股）
stock_code = '01810'
market_code = '116'  # 港股市场代码

# 构造请求URL和参数
url = 'http://push2his.eastmoney.com/api/qt/stock/kline/get'
params = {
    'secid': f'{market_code}.{stock_code}',
    'ut': 'fa5fd1943c7b386f172d6893dbfba10b',
    'fields1': 'f1,f2,f3,f4,f5,f6',
    'fields2': 'f51,f52,f53,f54,f55,f56,f57,f58,f59,f60,f61',
    'klt': '101',  # 日K线
    'fqt': '1',
    'beg': '20230101',
    'end': '20231231',
}

# 添加请求头
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Referer': 'http://quote.eastmoney.com/'
}

# 发送请求
response = requests.get(url, params=params, headers=headers)
data = response.json()

# 解析数据
klines = data['data']['klines']
rows = []
for kline in klines:
    # 分割字段：日期,开盘,收盘,最高,最低,成交量,成交额,振幅,涨跌幅,涨跌额,换手率
    items = kline.split(',')
    rows.append(items)

# 创建DataFrame
columns = ['日期', '开盘', '收盘', '最高', '最低', '成交量', '成交额', '振幅', '涨跌幅', '涨跌额', '换手率']
df = pd.DataFrame(rows, columns=columns)

# 转换数据类型
numeric_cols = ['开盘', '收盘', '最高', '最低', '成交量', '成交额', '振幅', '涨跌幅', '涨跌额', '换手率']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# 保存到CSV
df.to_csv('xiaomi_stock.csv', index=False, encoding='utf_8_sig')
print("数据已保存为 xiaomi_stock.csv")

# 数据可视化
plt.figure(figsize=(18, 6))
plt.plot(pd.to_datetime(df['日期']), df['收盘'], label='收盘价')
plt.title('小米集团-W 股票收盘价走势')
plt.xlabel('日期')
plt.ylabel('价格（HKD）')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('stock_price.png')
plt.show()
```
